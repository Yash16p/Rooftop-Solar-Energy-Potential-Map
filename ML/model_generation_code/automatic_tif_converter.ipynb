```python
import os
import tifffile
import imageio
import imagecodecs
```

```python

# Set the input and output folders
input_folder = 'dataset/source'
output_folder = 'dataset/output'

```

```python
# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

```

```python
# Loop through all files in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is a.tif file
    if filename.endswith('.tif'):
        # Open the.tif file using tifffile
        with tifffile.TiffFile(os.path.join(input_folder, filename)) as tif:
            image = tif.asarray()
        
        # Save the image as a.jpg file using imageio
        output_filename = os.path.join(output_folder, filename.replace('.tif', '.jpg'))
        imageio.imwrite(output_filename, image, format='jpg')

        print(f"Converted {filename} to {output_filename}")
```
